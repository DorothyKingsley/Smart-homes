```python
import bson
import pandas as pd
import pymongo
import csv
with open('/Users/dorothyjeyson/Downloads/SH-1/2019-03-28/sphere/ENV.bson','rb') as f:
    data1=bson.decode_all(f.read())
df=pd.DataFrame(data1)
df['bt_date'] = df['bt'].dt.date
df['bt_time'] = df['bt'].dt.time
df.drop(['bt','ts','tso','hid'], inplace=True, axis=1, errors='ignore')
#df.to_csv('ENV.csv', sep=',', encoding='utf-8',index=False)
df1=df['e']
#df1
variable_n=[]
for i in df1:
    for j in range(len(i)):
        for k,v in i[j].copy().items():
            if k=='n':
                variable_n.append(v)
a=list(set(variable_n))
print(a)
df
```

```python
final=[]
for i in df1:
    result={'BMP_PRES':'','HDC_TEMP':'','LT':'','ELEC':'','BMP_TEMP':'','PIR_TRIGS':'','HDC_HUM':''}
    for j in range(len(i)):
        for k,v in i[j].copy().items():
            col=i[j]['n']
            val=i[j]['v']
            result[col]=val
    final.append(result)
df2=pd.DataFrame(final)
```
